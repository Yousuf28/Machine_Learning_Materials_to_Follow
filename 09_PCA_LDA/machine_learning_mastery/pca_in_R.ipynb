```r
library(ggbiplot);
library(psych);
```

```r
# make fake data set
aa <- c(1,3,5,5,6,4,4,6,6,7,5,4,8,9,8,4,5,6,7,8)
AA <- matrix(aa, nrow = 5, ncol = 4)
AA
```

```r
# add column name
colnames(AA) <- c("location", "age", "neighbor", "area")
AA
```

```r
# add row name
rownames(AA) <- c("ob_01", "ob_02", "ob_03", "ob_04", "ob_05")
AA
```

## using prcomp function

```r
# pca with out scaling

pca_no_scale  <- prcomp(AA)
```

```r
# proportion variance comes from eigenvalues
summary(pca_no_scale)
```

```r
# all principle components (eigenvector)
## loading scores
pca_no_scale$rotation
```

$$ PC = loading.score * locatn + loading.score * age + loading.score * neigbor + loading.score *area $$

$$ PC1 = 0.55 * locatn + 0.38 * age + 0.58 * neigbor + 0.43 *area$$

```r
# sample
pca_no_scale$x
```

```r
ggbiplot(pca_no_scale)
```

```r
ggbiplot(pca_no_scale, labels=rownames(AA))
```

```r
pca_center <- prcomp(AA, center = TRUE)
```

```r
pca_center$rotation
```

## manual calculation

```r
AA
```

```r
#find mean of each column and make matrix so that I can substract from AA
mean <- colMeans(AA)
mean <- rep(mean, times = 1,each = 5)
mean <- matrix(mean, nrow = 5, ncol = 4)
mean
```

```r
# center , substract column mean from column
center <- AA - mean
center
```

```r
# covaraince 
center_cov <- cov(center)
center_cov
```

```r
# vectors, or principle component
pca_center_eigen  <- eigen(center_cov)
```

```r
#eigenvalues
pca_center_eigen$values
```

```r
# calculate proportion of variance from eignvalues

round((pca_center_eigen$values)/sum(pca_center_eigen$values), 4)
```

```r
summary(pca_center)
```

```r
# eigenvector or principle component
pca_center_eigen$vectors
```

```r
# PCs from prcomp function 
pca_center$rotation
```

## scale the feature 

```r
# from phych package
SD(AA)
```

```r
sd <- SD(AA)
sd <- rep(sd, times = 1,each = 5)
sd <- matrix(sd, nrow = 5, ncol = 4)
sd
```

```r
center_scale <- center /sd
center_scale
```

```r
# check everything fine
scale(AA)
```

```r
# when center scale , mean that standardize, actually z score use 
# it correlation
center_scale_cov <- cov(center_scale)
center_scale_cov
```

```r
# correlation
cor(center_scale)
```

```r
pca_center_scale_eigen <- eigen(center_scale_cov)
```

```r
pca_center_scale_eigen$vectors
```

```r
# from prcomp function using center and scale true (sign different only, does not matter)
pca_center_scale$rotation
```

```r
round((pca_center_scale_eigen$values)/ (sum(pca_center_scale_eigen$values)),4)
```

```r
summary(pca_center_scale)
```
